<span style="color:black;font-size:2em;font-weight:bold">3.</span> Exprimer le rendement locatif en fonction de $X$ et $Y$.

<span style="color:black;font-size:2em;font-weight:bold">4.</span> Définir la fonction python nommée <b>rendement</b> de paramètres <code style='background-color:white'>X</code> et <code style='background-color:white'>Y</code> qui renvoie le <b>rendement locatif</b>.

```python
def rendement(X, Y):
    revenus = 680 * (12 - X)
    frais = Y + (600 + 220 + 800 + 1000)
    return (revenus - frais) / 103000
```

<span style="color:black;font-size:2em;font-weight:bold">5.</span> Vérifier que le <b>rendement locatif</b> du bien  immobilier loué à temps plein et sans travaux est environ égal à $5,4\%$.

```python
rendement(0, 0)
```

<span style="color:black;font-size:2em;font-weight:bold">6.</span> Calculer le <b>rendement locatif</b> pour le bien loué à temps plein avec 500 € de travaux annuel.

```python
rendement(0, 500)
```

<h1 style="color:red;font-size:3em;font-weight:bold">Rendement moyen et Value at Risk</h1>

<span style="color:black;font-size:2em;font-weight:bold">6.</span> Charger la fonction <b>choices</b> du module <b>random</b>.

```python
from random import choices
```

<span style="color:black;font-size:2em;font-weight:bold">7.</span> Décrire ce que représente la variable <code style='background-color:white'>mois_sim</code> ci-dessous.

```python
mois_sim = choices(V, F1, k = 500000)
print(mois_sim)
```

<span style="color:black;font-size:2em;font-weight:bold">8.</span> Créer la liste <code style='background-color:white'>travaux_sim</code> renvoyant 500000 simulations de montants de travaux.

```python
travaux_sim = choices(T, F2, k = 500000)
```

<span style="color:black;font-size:2em;font-weight:bold">9.</span> Construire la liste de 500000 rendements locatifs simulés.

```python
L = [rendement(mois_sim[k], travaux_sim[k]) for k in range(500000)]
```

```python
pl.hist(L, bins = 20)
pl.show()
```

<div style="color:black;font-size:2em;font-weight:bold">Définition :</div> 
<blockquote style='border-left: 5px solid black;background-color:white'>Le  <b>rendement locatif moyen</b> est la moyenne d'un grand nombre de rendements locatifs simulés.
</blockquote>
